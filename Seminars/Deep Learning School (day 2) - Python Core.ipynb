## Classes

```python
class Human:
    def __init__(self, name='', age=None, deep_learning_specialist=False):
        self.name = name
        self.age = age
        self.deep_learning_specialist = deep_learning_specialist
        
    def set_age(self, age) -> int:
        self.age = age
        
    def get_age(self) -> int:
        return self.age
    
    def test(self) -> None:
        print("Aaa")
        
    def __str__(self) -> str:
        return f'Name: {self.name}, age: {self.age}'
```

```python
human = Human('Person', 17)
human
```

```python
human.test()
```

```python
human.get_age()
```

```python
print(human)  # выводит содержимое метода __str__
```

```python
class DLSchoolStudent(Human):  # этот класс наследуется от Human
    def __init__(self, name='', age=None, deep_learning_specialist=True):
        super().__init__(name, age)
        self.total_grade = None
        self.deep_learning_specialist = deep_learning_specialist
        
    def test(self) -> None:
        print('BBB')
```

```python
student = DLSchoolStudent('Good Person', 18)
print(student)
```

```python
student.test()
```

```python
student.deep_learning_specialist
```
